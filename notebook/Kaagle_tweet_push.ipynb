```python
#preprocessing data
from transformers import BertTokenizer
token = BertTokenizer.from_pretrained('bert-base-cased');
```

```python
X = df['tweet'].values
y = df['Toxicity'].values
```

```python
seqs = [seq for seq in X]
model_input = token(seqs, padding=True , return_tensors='pt')
```

```python
dataset = tf.data.Dataset.from_tensor_slices((model_input['input_ids'],y))
```

```python
#get partision for dataset into cache
dataset = dataset.cache()
dataset = dataset.shuffle(16000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

```python
#dataset divided
train = dataset.take(int(len(dataset)*.7))
test = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
val = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```
